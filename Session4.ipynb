```python
dfHer = pd.read_excel("Ressources/data-heritage-2019.xls", index_col=0)
dfHer.columns
```

```python
notAvailable = []
for index, name in df["Country (according to CN Gov):"].iteritems():
    requestedName = name
    if " (not bri)" in requestedName.lower() :
        requestedName = requestedName[:-10]
        #print("New name = " + name)

    if requestedName == "Thailiand" :
        requestedName = "Thailand"
    if requestedName == "The United Arab Emirates" :
        requestedName = "United Arab Emirates"        
    if requestedName == "United States of America" :
        requestedName = "United States"        
    if requestedName == "Kyrgyzstan" :
        requestedName = "Kyrgyz Republic"        
    if requestedName == "Myanmar" :
        requestedName = "Burma"       
    if requestedName == "Brunei" :
        requestedName = "Brunei Darussalam"    
    if requestedName == "East Timor":
        requestedName = "Timor-Leste"
            
    tempDfHer = dfHer[dfHer["Country Name"] == requestedName]
    try :
        rank = tempDfHer["World Rank"]
        stat = tempDfHer["2019 Score"]
        #print (isocode, rank, stat)
        df.at[index, "HF Index of Economic Freedom (Scale: 1-100)"] = stat
        df.at[index, "Heritage Foundation Index of Economic Freedom (Global Ranking)"] = rank
    except ValueError :
        notAvailable.append(name)
        continue
df[["Country (according to CN Gov):", 'HF Index of Economic Freedom (Scale: 1-100)', 'Heritage Foundation Index of Economic Freedom (Global Ranking)']]

if (len(notAvailable) == 0) :
    print ("Everything is fine")
else :
    print("No data for :")
    for ctry in notAvailable :
        print (ctry)
```

```python
dfRat = pd.read_csv("Ressources/data-ratings-2015.csv", sep= ";")
dfRat.columns
```

```python
iso_codes = pd.read_csv("Ressources/country-iso.csv", dtype="str")
def getiso2(iso3) :
    
    result = str(iso_codes[iso_codes["Alpha-3 code"] == iso3]["Alpha-2 code"])
    if result [5] == " " :
        if result [6] == " " :
            return result[7:9]
        else :
            return result[6:8]
    else :
        return result[5:7]
getiso2("ALB")
```

```python
dfRat2015 = dfRat[dfRat["Year"] == 2015]
notAvailable = []
for index, isocode in df.isocode.iteritems():
    #print(getiso2(isocode))
    try :
        tempDfRat = dfRat2015[dfRat2015["Country2Code"] == getiso2(isocode)]
    except KeyError :
        notAvailable.append(isocode)
        continue
    try :
        moodys = tempDfRat["Moodys_NumericRating"]
        sp = tempDfRat["SP_NumericRating"]
        fitch = tempDfRat["Fitch_NumericRating"]
        df.at[index, "Moody's Socvereign Ratings List"] = moodys
        df.at[index, "S&P Sovereign Ratings List"] = sp
        df.at[index, "Fitch Sovereign Ratings List"] = fitch
    except ValueError :
        notAvailable.append(isocode)
        continue
df[["Country (according to CN Gov):", "Moody's Socvereign Ratings List", "S&P Sovereign Ratings List", "Fitch Sovereign Ratings List"]]
if (len(notAvailable) == 0) :
    print ("Everything is fine")
else :
    print("No data for :")
    for ctry in notAvailable :
        print (ctry)
```
