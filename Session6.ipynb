```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("Ressources/UTSEUS-MOBIKE-shanghai_full.csv", index_col = "orderid")
df.head()
```

```python
minX = 121.10157 ;
maxX = 121.804611 ;
minY = 30.975 ;
maxY = 31.514999 ;

df = df[(df["start_location_x"] >= minX) & (df["start_location_x"] <= maxX) & (df["end_location_x"] >= minX) & (df["end_location_x"] <= maxX) & (df["start_location_y"] >= minY) & (df["start_location_y"] <= maxY) & (df["end_location_y"] >= minY) & (df["end_location_y"] <= maxY)]
```

```python
%matplotlib inline
df.plot.scatter(x = "start_location_x", y = "start_location_y", s = 0.01, c = "#FF000050", figsize =(15,15))
```

```python

df["startTime"] = pd.to_datetime(df["start_time"])
df["endTime"] = pd.to_datetime(df["end_time"])
df["duration"] = df["endTime"] - df["startTime"]
```

```python
#df = pd.read_pickle("mobike-df.pickle")
```

```python
from pyproj import CRS

mercator = CRS.from_epsg(4326)
china = CRS.from_epsg(4479)

from pyproj import Transformer

transformer = Transformer.from_crs(mercator, china, always_xy=True)
```

```python
#df[["start_location_x", "start_location_y"]].apply(lambda x : print(x[1]), axis = 1)
transformer.transform(df["start_location_x"][1], df["start_location_y"][1])
```

```python
df["transformed_start_location_x"] = df.apply(lambda x : transformer.transform(x["start_location_x"], x["start_location_y"])[0], axis = 1)
df["transformed_start_location_y"] = df.apply(lambda x : transformer.transform(x["start_location_x"], x["start_location_y"])[1], axis = 1)
df["transformed_end_location_x"] = df.apply(lambda x : transformer.transform(x["end_location_x"], x["end_location_y"])[0], axis = 1)
df["transformed_end_location_y"] = df.apply(lambda x : transformer.transform(x["end_location_x"], x["end_location_y"])[1], axis = 1)
```

```python
df.head()
```

```python
from scipy.spatial import distance

df["distance"] = df.apply(lambda x : distance.euclidean([x["transformed_start_location_x"], x["transformed_start_location_y"]], [x["transformed_end_location_x"], x["transformed_end_location_y"]]), axis = 1)
```

```python
df["mean_speed_in_kmh"] = df.apply(lambda x : x["distance"] / x["duration"].seconds * 3.6, axis = 1)
```
